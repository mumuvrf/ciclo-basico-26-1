## 📦 Preparação

Execute o código fornecido para carregar sua base personalizada.

```python
from pathlib import Path
import pandas as pd

BASE_DIR = Path.cwd()
NOME_TXT = BASE_DIR / "nome.txt"

if NOME_TXT.exists():
    nome = NOME_TXT.read_text(encoding="utf-8").strip()
else:
    nome = input("Digite seu nome: ").strip().lower().replace(" ", "")
    NOME_TXT.write_text(nome, encoding="utf-8")

caminho_csv = BASE_DIR / f"basetitanic{nome}.csv"

if not caminho_csv.exists():
    url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv"
    
    df_bruto = pd.read_csv(url)
    df_unico = df_bruto.sample(n=800).reset_index(drop=True)
    df_unico.to_csv(caminho_csv, index=False)

df = pd.read_csv(caminho_csv)

print(f"Base de {nome} carregada com sucesso!")
print(f'tamanho: {len(df)} linhas')
```
